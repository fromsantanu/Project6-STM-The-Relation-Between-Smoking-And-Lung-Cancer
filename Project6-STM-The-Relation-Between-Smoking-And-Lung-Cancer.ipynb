# Correlation between Smoking and chances of Lung Cancer

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Generate synthetic data
np.random.seed(0)
num_samples = 100
average_cigarettes_per_day = np.random.randint(0, 40, num_samples)
# Simulate lung cancer incidence (0 or 1) based on smoking, with less noise for clarity
lung_cancer_incidence = (average_cigarettes_per_day + np.random.normal(0, 2, num_samples) > 20).astype(int)

# Create a DataFrame
data = pd.DataFrame({
    'Average_Cigarettes_Per_Day': average_cigarettes_per_day,
    'Lung_Cancer_Incidence': lung_cancer_incidence
})

print("Sample Data used for the exercise")
print(data.head())
print()

# Calculate the correlation coefficient
correlation_coefficient, _ = pearsonr(data['Average_Cigarettes_Per_Day'], data['Lung_Cancer_Incidence'])
print(f"Correlation Coefficient calculated as : {correlation_coefficient:.2f}")
print("Data showing a strong positive correlation ..")
print()
print("Plot showing how Smoking and Lung Cancer are related")

# Plot the data
plt.figure(figsize=(10, 6))
plt.scatter(data['Average_Cigarettes_Per_Day'], data['Lung_Cancer_Incidence'], alpha=0.6, edgecolors='w', linewidth=0.5, c=data['Lung_Cancer_Incidence'], cmap='bwr')
plt.title('Correlation between Smoking and Lung Cancer Incidence')
plt.xlabel('Average Cigarettes Per Day')
plt.ylabel('Lung Cancer Incidence')
plt.xticks(np.arange(0, 41, 5))
plt.yticks([0, 1], ['No Cancer', 'Cancer'])

# Add a trendline
z = np.polyfit(data['Average_Cigarettes_Per_Day'], data['Lung_Cancer_Incidence'], 1)
p = np.poly1d(z)
plt.plot(data['Average_Cigarettes_Per_Day'], p(data['Average_Cigarettes_Per_Day']), color='red')

plt.colorbar(label='Lung Cancer Incidence')
plt.show()
```
